**Australian Brushturkey_sound**

```python
from pydub import AudioSegment
import os

# Initialize an empty AudioSegment to hold the concatenated audio
concatenated_audio = AudioSegment.empty()

# Specify the directory path where your audio files are located
audio_folder = '/kaggle/input/sound-of-114-species-of-birds-till-2022/Voice of Birds/Voice of Birds/Australian Brushturkey_sound'

# List all the audio files in the specified folder
audio_files = [os.path.join(audio_folder, file) for file in os.listdir(audio_folder) if file.endswith('.mp3')]

# Loop through the audio files and concatenate them
for file_path in audio_files:
    audio_segment = AudioSegment.from_mp3(file_path)  # Load the audio file
    concatenated_audio += audio_segment  # Concatenate the audio data

# Specify an absolute path for the output file with .wav extension
output_path = '/kaggle/working/Australian_Brushturkey_sound/Australian_Brushturkey_sound.wav'

# Create the directory if it doesn't exist
output_directory = os.path.dirname(output_path)
os.makedirs(output_directory, exist_ok=True)

# Export the concatenated audio to a temporary .wav file
temp_wav_path = '/kaggle/working/Australian_Brushturkey_sound/temp_audio.wav'
concatenated_audio.export(temp_wav_path, format="wav")

# Convert the .wav to .mp3 using pydub (if needed)
converted_audio = AudioSegment.from_wav(temp_wav_path)

# Save the converted audio as .wav
converted_audio.export(output_path, format="wav")

# Clean up the temporary .wav file
os.remove(temp_wav_path)

print(f"Concatenated audio saved as {output_path}")
```

```python
from pydub import AudioSegment
import os

# Specify the path to the concatenated audio file (in .wav format)
concatenated_audio_path = '/kaggle/working/Australian_Brushturkey_sound/Australian_Brushturkey_sound.wav'

# Load the concatenated audio file
concatenated_audio = AudioSegment.from_wav(concatenated_audio_path)

# Calculate the duration of each split (3 seconds)
split_duration = 3 * 1000  # 3 seconds in milliseconds

# Define the output directory for the split files
output_directory = '/kaggle/working/Australian_Brushturkey/'

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Split the audio into 3-second segments
for i, start_time in enumerate(range(0, len(concatenated_audio), split_duration)):
    end_time = start_time + split_duration
    split = concatenated_audio[start_time:end_time]
    
    # Define the output file path
    output_path = os.path.join(output_directory, f'Australian_Brushturkey{i + 1}.wav')
    
    # Export the split segment to a .wav file
    split.export(output_path, format="wav")

    print(f"Australian_Brushturkey {i + 1} saved to {output_path}")
```
